## Input Embedding

### One-Hot Encoding

在CV中，我们通常将输入图片转换为4维(batch, channel, height, width)张量来表示；而在NLP中，可以将输入单词用One-Hot形式编码成虚列向量，向量长度为预定义的词汇中拥有的单词量，向量在这一维中的值只有一个位置为1，其他都为0，1对应的位置就是词汇表中表示这个词的地方。

### Word Embedding

One-Hot形式上很简洁，但劣势在于太稀疏，而且可能很长。比如词汇表如果有10k个词，那么一个词向量的长度就需要达到10k，而其中却仅有一个为1，其余全为0，太浪费！

更重要的是，这种方法无法体现出词与词之间的关系，比如‘爱’和‘喜欢’这两个词，他们的意思是相近的，但基于One-hot编码后的结果取决于他们在词汇表中的位置，无法体现出他们之间的关系。

因此我们需要另一种词的表示方法，能够体现词与词之间的关系，使意思相近的词有相近的表达结果，这种方法即Word Embedding.

那么应该如何设计这种方法呢？最方便的途径是设计一个可学习的权重矩阵W，将词向量与这个矩阵进行点乘，即得到新的结果。
