#### Channels

channels:the path or the location where conda looks for the packages that we want to install.  



- channel:`conda config --show channels`

- channel:`conda install -c pytorch pytorch` (pytorch channelchannel list) 

- channel condaforge.com `conda config --add channels conda-forge`      
    - condaforge.com :A community-led collection of recipes, build infrastructure and distributions for the conda package manager

- channelchannel:`conda config -add channels pytorch`  
                              `conda install pytorch`      
                              `conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free/`   
                              `conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/main/`  
                              `conda config --set show_channel_urls yes`
