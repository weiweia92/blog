## 图片读写及显示

```python
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('pic/1.jpeg')
```

```python
cv.imwrite('test/test.jpg', img)
```

```python
plt.imshow(img)
plt.show()  #GBR
```

```python
# RGB-->GBR
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
```

```python
img3 = cv.imread('pic/1.jpeg', 0)
plt.imshow(img3)
plt.show()
```

```python
plt.imshow(img3, cmap='gray')
plt.show()
```

```python
img.ndim
```

```python
img3.ndim
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
img4 = cv.imread('pic/2.jpeg')
show(img4)
```

```python
gray = cv.imread('pic/2.jpeg', 0)
show(gray)
```

## 简单绘图

```python
x = np.arange(2, 20)
y = 2 * x + np.random.randint(5, 20, 18)
```

```python
plt.plot(x, y, '*', color='r')
plt.show()
```

```python
x = np.linspace(0, 1, 100)
y1 = np.power(x, 0.5)
y2 = x
y3 = np.power(x, 1.5)
```

```python
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
plt.plot(x, y1, label='0.5')
plt.plot(x, y2, label='1.0')
plt.plot(x, y3, label='1.5')
plt.legend()
plt.xlabel('r')
plt.ylabel('s')
plt.grid()   #加网格
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.show()
```
