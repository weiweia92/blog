```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

## 两图像的加减乘除

```python
obj = cv.imread('pic/cartoon.jpeg', 0)
bg = cv.imread('pic/bk.jpeg', 0)
mask = cv.imread('pic/mask.jpeg', 0)
noise = cv.imread('pic/noise.jpeg', 0)
```

```python
obj.shape
```

```python
bg.resize((225, 225))
```

```python
bg.shape
```

```python
mask.shape
```

```python
noise.shape
```

```python
show(np.hstack([obj, bg, mask, noise]))
```

```python
img_add = cv.add(obj, bg)   #相加元素大于255就截断为255， 为解决这个问题通常对图片加权重
show(img_add)
```

```python
img_add1 = cv.add(obj*0.5, bg*0.5)
```

```python
show(img_add1)
```

```python
img_add2 = cv.addWeighted(obj, 0.5, bg, 0.5, 0) #0偏移值
show(img_add2)
```

```python
img_add1.dtype, img_add2.dtype
```

```python
#减法
img_sub = img_add1 - bg * 0.5
show(img_sub)
```

```python
img_sub.max()
```

```python
img_sub = 2*(img_add1 - bg * 0.5)
show(img_sub)
```

```python
img_sub.max()
```

```python
img_sub2 = cv.subtract(img_add1, bg*0.5)
show(np.hstack([img_sub, img_sub2]))
```

```python
img_sub2.max()
```

```python
#乘法
mask.max()  
```

```python
mask = mask / 255
```

```python
img_mul = obj * mask
img_mul2 = cv.multiply(obj*1.0, mask)
show(np.hstack([img_mul, img_mul2]))
```

```python
#除法
img_div = obj / (np.float64(noise)+1)
img_div2 = cv.divide(obj, noise+1)
show(np.hstack([img_div, img_div2]))
```

## 图像的线性和非线性变换

$ s = b+kr $  
$ s = a + \frac{ln(r+1)}{b}  $     
$ s = cr^{\gamma}$

```python
img = cv.imread('pic/cat.jpeg', 0)
show(img)
```

```python
b = 20
k = 2
img2 = b + k * img  #img2中的数据大于255会进行0-255的循环产生奇怪的数
show(img2)
```

```python
img3 = b + k * img.astype(np.int32)  #将img3类型转换成可以表示比较大的数字的值
show(img3)
```

```python
img3.max()   
```

```python
img3 = np.clip(0, 255, img3)
show(img3)
```

```python
img4 = cv.convertScaleAbs(img, alpha=2, beta = 20)
show(img4)
```

```python
img5 = 10 + np.log(img.astype(np.float32) + 1)/ 0.1
show(img5)
```

```python
img01 = img / 255
img05 = np.power(img01, 0.5) * 255
img15 = np.power(img01, 1.5) * 255
show(np.hstack([img05, img, img15]))
```

## 图像融合

```python
A = cv.imread('pic/cartoon.jpeg')
B = cv.imread('pic/noise.jpeg')
C = cv.imread('pic/mask.jpeg')
```

```python
A.shape, B.shape, C.shape
```

```python
show(B)
```

```python
show(np.hstack([A, B, C]))
```

```python
C = C / 255
As = A * (1 - C)
Bs = B * C
show(np.uint8(np.hstack([As, Bs])))
```

```python
nice = 0.1*As + 0.9*Bs
show(nice.astype(np.uint8))
```
