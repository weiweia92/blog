### 空间距离卷积核
$$ c(\xi-x)=e^{-\frac{||\xi-x||^2}{2\sigma^2_d}} $$
### 灰度差异卷积核
$$ s(f(\xi)-f(x))=e^-\frac{||f(\xi)-f(x)||^2}{2\sigma^2_r} $$

```python
def get_C(sigmad, n): #n指卷积核大小
    C = np.zeros((n, n))
    x = np.array([n//2, n//2])
    
    for i in range(n):
        for j in range(n):
            ksi = np.array([i, j])
            C[i,j] = np.exp(-0.5 * (np.linalg.norm(ksi - x)/ sigmad)**2)
            
    C /= C.sum()  #C 进行归一化
    return C
```

```python
C = get_C(3, 11)
show(C)
```

```python
def get_S(f, sigmar, n):
    S = np.zeros((n, n))
    
    f = np.float64(f)
    for i in range(n):
        for j in range(n):
            S[i, j] = np.exp(-0.5 * ((f[i, j] - f[n//2, n//2]) / sigmar) **2)
    S /= S.sum()
    return S
```

```python
def get_S2(f, sigmar, n):
    f = np.float64(f)
    
    S = np.exp(-0.5 * ((f-f[n//2,n//2]) / sigmar)**2)
    S /= S.sum()
    return S
```

```python
S1 = get_S(C, 10, 11)
show(S1)
```

```python
S2 = get_S(C, 10, 11)
show(S2)
```

```python
sigmar = 50
sigmad = 3
n = 11

h, w = img.shape
img2 = np.zeros_like(img)
```

```python
%%time
C = get_C(sigmad, n)
for i in range(h-n):
    for j in range(w-n):
        f = img[i:i+n, j:j+n]
        S = get_S2(f, sigmar, n)
        K = C * S
        K /= K.sum()
        img2[i, j] = (f * K).sum()

show(img2)
```
