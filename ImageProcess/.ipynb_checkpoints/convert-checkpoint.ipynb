```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

## 图像的裁剪，放大和缩小

```python
img = cv.imread('pic/cat.jpeg')
show(img)
```

```python
img.shape
```

```python
cat = img[30:200, 50:250, :]
```

```python
show(cat)
```

```python
cat.shape
```

```python
img2 = cv.resize(img, (500, 300))
show(img2)
```

```python
#插值方法：缩小时用INTER_AREA, 放大时用INTER_LINEAR或INTER_NEAREST
img3 = cv.resize(img, (500, 400), interpolation=cv.INTER_NEAREST)
show(img3)
```

## 平移变换

```python
show(img)
```

```python
img.shape
```

```python
M = np.array([
    [1, 0, 100],
    [0, 1, 50]
], dtype=np.float32)  #水平平移100，竖直平移50
```

```python
img2 = cv.warpAffine(img, M, (500, 300))
show(img2)
```

## 错切变换

```python
M = np.array([
    [1, 0.2, 0],
    [0, 1, 0]
], dtype=np.float32)
```

```python
img3 = cv.warpAffine(img, M, (300, 212))
show(img3)
```

```python
M = np.array([
    [1, 0, 0],
    [0.3, 1, 0]
], dtype=np.float32)
```

```python
img4 = cv.warpAffine(img, M, (238, 300))
show(img4)
```

## 镜像变换

```python
Mx = np.array([
    [-1, 0, 238],
    [0, 1, 0]
], dtype=np.float32)
img5 = cv.warpAffine(img, Mx, (238, 212))
show(img5)
```

```python
My = np.array([
    [1, 0, 0], 
    [0, -1, 212]
], dtype=np.float32)
```

```python
img6 = cv.warpAffine(img, My, (238, 212))
show(img6)
```

```python
img7 = cv.flip(img, 1)
img8 = cv.flip(img, 0)
img9 = cv.flip(img, -1)
```

```python
show(np.hstack([img7, img8, img9]))
```

## 旋转图片

```python
beta = np.pi / 4
M = np.array([
    [np.cos(beta), np.sin(beta), 0],
    [-np.sin(beta), np.cos(beta), 0]
], dtype=np.float32)
```

```python
img10 = cv.warpAffine(img, M, (400, 212))
show(img10)
```

```python
h, w, c = img.shape
```

```python
M2 = cv.getRotationMatrix2D((w//2, h//2), 45, 1)
M2
```

```python
img11 = cv.warpAffine(img, M2, (238, 212))
show(img11)
```

img12 = cv.rotate(img, cv.ROTATE_90_CLOCKWISE) #顺时针旋转90度

```python
show(img12)
```

## 透视变换

```python
img = cv.imread('pic/parthenon.jpeg')
show(img)
```

```python
src = np.array([
    [140, 35],
    [230, 75],
    [240, 150],
    [135, 140]
], dtype=np.float32)
```

```python
dst = np.array([
    [74, 15],
    [250, 30],
    [250, 140], 
    [75, 150]
], dtype=np.float32)
```

```python
M = cv.getPerspectiveTransform(src, dst)
M
```

```python
h, w, c = img.shape
```

```python
img2 = cv.warpPerspective(img, M, (w, h))
show(img2)
```
