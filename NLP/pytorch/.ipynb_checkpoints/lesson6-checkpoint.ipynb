## 一.AutoGrad的回溯机制与动态计算图

### 1.可微分性相关属性

PyTorch中的张量已经不仅仅是一个纯计算的载体，张量本身也可支持微分运算。这种可微分性不仅体现在我们可以使用grad函数进行求导，更重要的是这种可微分性会体现在可微分张量参与所有运算中。

* requires_grad属性：可微分性

```python
# 构建可微分张量
x = torch.tensor(1., requires_grad=True)
x
```

```python
# 构建函数关系
y = x ** 2
```

* grad_fn属性：存储Tensor微分函数

```python
y
```

我们发现此时张量y具有一个`grad_fn`属性，并且取值为`<PowBackward0>`,我们可以查看该属性

```python
y.grad_fn
```

`grad_fn`其实是存储了Tensor的微分函数，或者说`grad_fn`存储了可微分张量在进行计算的过程中函数关系，此处x到y其实就是进行了幂运算。

```python
# 但x作为初始张量，并没有grad_fn属性
x.grad_fn
```

这里值得注意的是，y不仅和x存在幂运算关系，更重要的是，y本身还是一个由x张量计算得到的一个张量。

```python
y
```

而对于一个可微分张量生成的张量，也是可微分的

```python
y.requires_grad
```

相比于x,y不仅同样拥有张量的取值，并且同样可微，还额外存储了x到y的函数计算信息。我们再尝试围绕y创建新的函数关系z = y + 1

```python
z = y + 1
```

```python
z
```

```python
z.requires_grad
```

```python
z.grad_fn
```

不难发现，z也同时存储了张量计算数值、z是可微的，并且z还存储了和y的计算关系(add)。据此我们可以知道，在PyTorch的张量计算过程中，如果我们设置初始张量是可微的，则在计算过程中，每一个由原张量计算得到的新张量都是可微的，并且还会保存此前一步的函数关系，也就是所谓的回溯机制。而根据这个回溯机制，我们能非常清楚的掌握张量的每一步计算，并据此绘制张量计算图。

### 2.张量计算图

借助回溯机制，我们就能够将张量的复杂计算过程抽象为一张图(Graph),例如此前我们定义的x、y、z三个张量，三者的计算关系就可以由下图进行表示。



* 计算图的定义

上图就是用于记录可微分张量计算关系的张量计算图，图由节点和有向边构成，其中节点表示张量，边表示函数计算关系，方向则表示实际运算方向，张量计算图本质是有向无环图。

* 节点类型

在张量计算图中，虽然每个节点都表示可微分张量，但是节点和节点之间却略有不同。就像在前例中，y和z保存了函数计算关系，但x没有，而在实际计算关系中，我们不难发现z是所有计算的终点，因此，虽然x、y、z都是节点，但每个节点却并不一样。此处我们可以将节点分为三类，分别是：   

a):叶节点，也就是初始输入的可微分张量，前例中x为叶子节点；    
b):输出节点，也就是最后计算得到的张量，前例中z为输出节点；     
c):中间节点，在一张计算图中，除了叶子节点和输出节点外其他都是中间节点，前例中y是中间节点。     
当然，在一张计算图中，可以有多个叶节点和中间节点，但大多数情况下，只有一个输出节点，若存在多个输出结果，我们也往往会将其保存在一个张量中。

### 3.计算图的动态性

值得一提的是，PyTorch的计算图是动态计算图，会根据可微分张量的计算过程自动生成，并且伴随着新张量或运算的加入不断更新，这使得PyTorch的计算图更加灵活高效，并且更加易于构建，相比于先构建图后执行计算的部分框架(老版本的TensorFlow),动态图更适用于面向对象编程。

## 二.反向传播与梯度计算

### 1.反向传播基本过程

首先，对于某个可微分张量的导数值(梯度值)，存储在grad属性中。

```python
x.grad
```

在最初，x.grad属性是空值，不会返回任何结果，我们虽然已经构建了x、y、z三者之间的函数关系，x也有具体取值，但要计算x点导数，还需要进行具体的求导运算，也就是执行所谓的反向传播。所谓反向传播可以简单理解为，在此前记录的函数关系基础上，反向传播函数关系，进而求的叶节点的导数值。在必要时求导，这也是节省计算资源和存储空间的必要规定。

```python
z
```

```python
z.grad_fn
```

```python
#执行反向传播
z.backward()
```

反向传播结束后，即可查看叶节点的导数值

```python
x
```

```python
#在z=y+1=x**2+1函数关系基础上，x取值为1时的导数值
x.grad
```

注意：在默认情况下，在一张计算图上执行反向传播，只能计算一次，再次调用backward方法将会报错

```python
z.backward()
```

在一张计算图上执行反向传播，只能计算一次，再次调用backward方法将会报错,所以我们又重新定义了一下计算图

当然，在y上也执行了反向传播

```python
x = torch.tensor(1., requires_grad=True)
y = x ** 2
z = y + 1
```

```python
y.backward()
```

```python
x.grad
```

```python
y.backward() #再次执行也会报错
```

```python
z.backward()
```

无论如何，我们只能计算叶节点的导数值

```python
y.grad
```

### 2.反向传播运算注意事项

* 中间节点反向传播和输出节点反向传播区别

尽管中间节点也可以进行反向传播，但很多时候由于存在复合函数关系，中间节点反向传播的计算结果和输出节点反向传播输出结果并不相同。

```python
x = torch.tensor(1., requires_grad=True)
y = x ** 2
z = y ** 2
```

```python
z.backward()
```

```python
x.grad
```

```python
x = torch.tensor(1., requires_grad=True)
y = x ** 2
z = y ** 2
```

```python
y.backward()
```

```python
x.grad
```

* 中间节点的梯度保存

默认情况，在反向传播过程中，中间节点并不会保存梯度

```python
x = torch.tensor(1., requires_grad=True)
y = x ** 2
z = y ** 2
```

```python
z.backward()
```

```python
y.grad
```

```python
x.grad
```

若想保存中间节点的梯度，我们可以使用retain_grad()方法

```python
x = torch.tensor(1., requires_grad=True)
y = x ** 2
y.retain_grad()
z = y ** 2
```

```python
z.backward()
```

```python
y
```

```python
y.grad
```

```python
x.grad
```

### 3.阻止计算图追踪

在默认情况下，只要初始张量时可微分张量，系统就会自动追踪其相关运算，并保存在计算图关系中，我们也可通过grad_fn来查看记录的函数关系，但在特殊情况下，我们并不希望可微张量从创建到运算结果输出都被记录，此时就可以使用一些方法来阻止部分运算被记录。

* `with torch.no_grad()`:阻止计算图记录

例如我们希望x、y函数关系被记录，而y的后续其他运算不被记录，可以使用with torch.no_grad()来阻止部分y的运算不被记录。

```python
x = torch.tensor(1., requires_grad=True)
y = x ** 2
```

```python
with torch.no_grad():
    z = y ** 2
```

with相当于是一个上下文管理器，with torch.no_grad()内部代码都"屏蔽"了计算图的追踪记录

```python
z
```

```python
z.requires_grad  #z不是可微分张量，也就不会纳入到计算图中，也就不会保存grad_fn的关系，.grad等
```

```python
y
```

* `.detach()`:创建一个不可导的相同张量

在某些情况下，我们可以创建一个不可导的相同张量参与后续运算，从而阻断计算图的追踪

```python
x = torch.tensor(1., requires_grad=True)
y = x ** 2
y1 = y.detach()
z = y1 ** 2
```

```python
y
```

```python
y1
```

```python
z
```

### 4.识别叶节点

由于叶节点较为特殊，如果需要识别在一个计算图中某张量是否是叶节点，可以使用is_leaf属性查看对应张量是否是叶节点

```python
x.is_leaf
```

```python
y.is_leaf
```

但is_leaf方法也有容易混淆的地方，对于任何一个新创建的张量，无论是否可导、是否加入计算图，都可以是叶节点，这些节点距离真正的叶节点，只差一个requires_grad属性调整。

```python
torch.tensor([1]).is_leaf
```

```python
# 经过detach的张量也可以是叶节点
y1
```

```python
y1.is_leaf
```

## 三.梯度下降基本思想

有了AutoGrad模块中函数方法的支持，我们可以尝试手动构造另一个优化算法：梯度下降算法。



### 2.梯度下降核心思想

此时a,b是在实数域上取值。假设二者初始值为0，也就是初始随机点为原点。对于(0, 0)点，有梯度计算如下：

```python
a = torch.tensor(0., requires_grad=True)
a
```

```python
b = torch.tensor(0., requires_grad=True)
b
```

```python
s0 = torch.pow((2 - a - b), 2) + torch.pow((4 - 3 * a - b), 2)
```

```python
s0
```

```python
s0.backward()
```

```python
a.grad, b.grad
```

也就是原点和(-28, -12)这个点之间连成直线的方向，就是能够使得SSE变化最快的方向，并且朝向(-28, -12)方向就是使得SSE增加最快的方向，反方向则是令SSE减少最快的方向。

```python
# 通过绘制直线，确定原点的移动方向
import matplotlib.pyplot as plt
x = np.arange(-30, 30, 0.1)
y = (12/28) * x
plt.plot(x, y, '-')
plt.plot(0, 0, 'ro')
plt.plot(-28, -12, 'ro')
```

逆梯度值的方向是使得SSE变小的最快方向，我们尝试移动‘一小步’。一步移动到（28， 12）是没有意义的。由于a和b的取值要按照(28, 12)等比例变化，因此我们不妨采用如下方法进行移动：
$$\left[
 \begin{matrix}
   0\\
   0
  \end{matrix} 
\right] + 0.01 * \left[
 \begin{matrix}
   28\\
   12
  \end{matrix} 
\right] = \left[
 \begin{matrix}
   0.28\\
   0.12
  \end{matrix} 
\right]$$


```python
s0
```

```python
a = torch.tensor(0.28, requires_grad=True)
a
```

```python
b = torch.tensor(0.12, requires_grad=True)
b
```

```python
s1 = torch.pow((2 - a - b), 2) + torch.pow((4 - 3 * a - b), 2)
s1
```

确实有所下降，继续求解新的点的梯度

```python
s1.backward()
```

```python
a.grad, b.grad
```
