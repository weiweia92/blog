At first glance, the only difference between the two datasets seems to be that dataset B is linearly separable, whereas dataset A is not.



It seems that the data is perfectly linearly separable in the case of dataset B, however this is not the case for dataset A.

Logistic regression attempts to maximize the log-likelihood:

$$\ell(\theta)=\sum_{i=1}^m y^i h_{\theta}(x^i)+(1-y^i)(1-h_{\theta}(x^i))$$

Where $h_{\theta}(x)=\frac{1}{1+\text{exp}(-\theta^T x)}$ is the sigmoid function.

The problem is a scalling issue: Since the data is linearly separable, $\parallel \theta \parallel$ can be made arbitrarily large and this will increase the log-likelihood indefinitely, as  $\forall i, h_{\theta}(x^i) \rightarrow 1 \text{when} \parallel \theta \parallel \rightarrow \infty$.This happens so long as $\theta$ describes a hyperplane that separates the data perfectly.

Another way of saying this is that $\Delta_{\theta}=X^T(Y-h_{\theta}(X))$ can be made arbitrarily close to 0 by increasing $\theta$, so long as $\theta$ describes a hyperplane that separates the data perfectly.

This does not happen when the data is not linearly separable: in this case, the log-likelihood would start decreasing due to the misclassified data points, i.e.  not $\exists \theta$ such that $X^T(Y-h_{\theta}(X))=0$


#### (c)

* **Using a different constant learning rate**

No. A different fixed learning rate is only a scalar for $\nabla_\theta J(\theta)$.

* **Decreasing the learning rate over time**

No. same as above: This is a scaling issue, changing the learning rate would not solve the problem

* **Adding a regularization term $\parallel \theta \parallel^2$ to the loss function**

Yes.This would solve the problem, as we would be actively penalizing large values of $\theta$ by including a term proportional to $\parallel \theta \parallel^2$  in the objective function.

* **Linear scaling of the input features**

This would not solve the problem, as any linear transformation would still keep the data linearly separable

* **Adding zero-mean Gaussian noise to the training data or labels**

This could work, as the data could be made to be linearly inseparable with the addition of noise. However, this is not a very robust solution as convergence of the algorithm depends on the set of noise samples: there is no guarantee the algorithm would converge in general for all possible sets of noise samples.

#### (d)

Support Vector Machines maximize the geometric margin. In the case of a linearly separable dataset, an SVM would maximize the distance between the separator and the closest data point, such that all data points are correctly classified. An SVM would therefore be immune to datasets such as B, as geometric margin is independant of $\parallel \theta \parallel$.
