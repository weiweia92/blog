高中数学有一个著名的概率问题，“一枚硬币连续投了五次都是正面，那么第六次投还是正面的概率是多少？”机智的高中生会想，这骗得了我？这是独立重复实验，概率还是0.5！

可如果一枚硬币连续投了一百次都是正面呢？一亿次都是正面呢？是否还要坚信出现正面的概率是0.5？这个硬币会不会被人动了手脚，就只有正面呢？

当大量的事实摆在我们面前，随着数据越来越多，人应该越来越相信这个硬币有问题，越来越不相信硬币的概率是0.5，这才是最自然而然的感觉，这才是动态的看待问题，而不是机械僵化的看待问题，而这种直觉背后就是贝叶斯思想。

现在，我们用下面这张图来讨论贝叶斯眼中的概率是如何变化的。一开始，还没有投掷硬币，我们对硬币一无所知，不如假设硬币的概率是0-1的均匀分布（如图左上）。然后，我们投掷了一次硬币，观察到1次正面，这时候我会想正面出现了，这个硬币是能得到正面，而反面还没出现，那我觉得这枚硬币的概率应该往1方向偏移一些（如图右上）。之后，我又投掷了10次，10次全是正面，我又调整了我对这枚硬币的看法，我对硬币正面概率靠近1的信心更强了（如图左下）。投掷了100次，100次全是正面，一次反面都没有，我会觉得硬币正面概率为1是“铁证如山”（如图右下）。



这时候，要是高中生和我说，这是独立重复实验，得到正面的概率应该是0.5。我会立马反驳，凭什么啊！你有证据吗？你有观测到反面吗？

下面，我们来看看正常硬币的实验过程应该是怎样的。一开始假设0-1均匀分布，然后观察到1次正面，我的信念往1偏一些，然后投掷10次，观察到6次正面，4次反面，10次里面出现6次，概率为0.6的可能性最大吧。最后，投掷100次，观察到53次正面，概率为0.53的可能性最大，随着数据的增加，我对概率估计的信心越来越强～



高中老师告诉我们概率就是频率；而贝叶斯告诉我们，概率其实有点像人对事物的信心。随着证据越来越多，我们对事物的信心越来越强，也可以称作概率。



这就是贝叶斯的思想，而AI的故事，将从这里开始。
