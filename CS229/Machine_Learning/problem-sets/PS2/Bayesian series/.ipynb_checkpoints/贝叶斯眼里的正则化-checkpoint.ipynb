前面讲到似然函数 $p(\text{data}|\theta)$ 是从掷骰子来的，所以：

$$P(\text{data}|\theta)=\theta^a (1-\theta)^b $$

其中得到正面 $a$ 次，反面 $b$ 次。

其实似然函数 $P(\text{data}|\theta)$ 不仅仅局限于此，掷骰子符合伯努利分布，所以似然函数的形式长上面那个样子。其他模型一样有似然函数，只要模型的输出是一个基于数据的概率，都可以被看做似然函数。如线性回归，逻辑回归，全连接神经网络DNN，卷积神经网络CNN，循环神经网络RNN，LSTM的输出经过softmax之后被转化为概率，全部都可以看作是似然函数 $P(\text{data}|\theta)$，只不过权重 $w$，偏差 $b$ 全部都包含到 $\theta$ 里去了，$P(\text{data}|\theta)$ 可以理解为在这样一组参数 $\theta$ 下，基于观测到数据所得到的概率。

那么从这个视角再来看什么是最大似然估计(Maximum Likelihood Estimation)，简称MLE。那么MLE对 $\theta$ 的估计方法可以写成如下形式：

$$\theta_{MLE}=\text{argmax}P(\text{data}|\theta)=\text{argmax}P(x_1|\theta)P(x_2|\theta)...P(x_n|\theta)=\text{argmax}\prod_{i=1}^n P(x_i|\theta)$$

两边同时取 $\text{log}$:

$$\theta_{MLE}=\text{argmax}\sum_{i=1}^n \text{log}P(x_i|\theta)$$

再同时取负，最大化问题转化为最小化问题：

$$\theta_{MLE}=\text{argmin}-\sum_{i=1}^n\text{log}P(x_i|\theta)$$

这就是Negative Log Likelihood(NLL)函数，是不是似曾相识，深度学习做分类任务时所用的cross entropy loss，其本质就是最大似然估计MLE!

那贝叶斯眼里的最大后验证估计(Maximum A Posterior Estimation,简称 MAP)是什么呢？那么最大后验证估计对 $\theta$ 的估计方法如下：

$$\theta_{MAP}=\text{argmax}P(\theta|\text{data})$$

两边同时取 $\text{log}$，取负，再用贝叶斯公式展开：

$$\theta_{MAP}=\text{argmin}-\text{log}P(\theta|\text{data})=\text{argmin}-\text{log}P(\text{data}|\theta)-\text{log}P(\theta)+P(\text{data})$$

$P(\text{data})$ 是一个固定的常数可以忽略，所以：

$$\theta_{MAP}=\text{argmin}-\text{log}P(\text{data}|\theta)-\text{log}P(\theta)$$

上式里 $-\text{log}P(\text{data}|\theta)$ 就是上面用 MLE 得到的 Negative Log Likelihood，所以最大似然估计和最大后验估计的区别就在于先验 $-\text{log}P(\theta)$。

那先验 $P(\theta)$ 到底是什么呢？字面上看是参数 $\theta$ 的概率分布，可以理解为你觉得参数应该怎么分布。我们先来假设参数符合高斯分布：

$$P(\theta) \sim e^{-\frac{\theta^2}{2\sigma^2}}$$

把上式带入到 $\theta_{MAP}$:

$$\theta_{MAP}=\text{argmin}-\text{log}P(\text{data}|\theta)-\text{log}e^{-\frac{\theta^2}{2\sigma^2}}=\text{argmin}-\text{log}P(\text{data}|\theta)+\frac{\theta^2}{2\sigma^2}$$

然后你会惊奇的发现用高斯分布作为先验的MAP等价于在MLE中加上L2的正则项！

此时我们便引入了一个新的角度来看正则项，即你觉得这个模型的参数分布应该是什么样子的。下面就是实际训练出来模型的参数分布，是不是看起来真的很高斯！加了正则项之后，其实就是给参数加了一个“紧箍咒”，告诉他说，我觉得参数分布应该是高斯分布的，你不要跑的太远哦～



那么还有没有其他先验呢？有啊，例如拉普拉斯分布。拉普拉斯分布就是下面的蓝线。其具体公式为：

$P(\theta)\sim \text{exp}\big(-\frac{|\theta|}{b}\big)$，把它带入到 $\theta_{MAP}$ 得：

$$\theta_{MAP}=\text{argmin}-\text{log}P(\text{data}|\theta)-\text{log}e^{-\frac{|\theta|}{b}}=\text{argmin}-\text{log}P(\text{data}|\theta)+\frac{\theta}{b}$$

你看拉普拉斯先验在０的位置具有最高的概率密度，这就说明很多参数会变成０，这就是为什么L1正则化会具有能产生稀疏性的优良性质。



总结一下，交叉熵损失函数（cross entropy loss），其本质就是最大似然估计MLE，而正则化等价于MLE加上先验分布。所以，从贝叶斯角度来看，损失函数＋正则化就是贝叶斯最大后验估计MAP。
