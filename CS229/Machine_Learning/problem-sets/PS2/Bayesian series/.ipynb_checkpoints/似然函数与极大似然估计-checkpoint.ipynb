

在动图中我们看到随着数据的增加，我们对硬币概率的估计逐渐发生变化。而在动图里的曲线是什么呢？

这个曲线是我们根据收集到的数据产生的似然函数：

$$p(\text{data}|\theta)=\prod_{n=1}^N p(x_n|\theta)\quad \text{data}=(x_1,x_2,...,x_n)$$

data表示数据的集合，包含了 $x_1$ 到 $x_n$ 个数据，就是每次掷硬币得到的正面反面，可以记作0和1或tail 和 head。$\theta=(p_1,...,p_n)$ 是参数的集合，例如掷硬币就只有正面反面两个概率 $p_1, p_2$，掷骰子就有六个面的六个概率 $p_1, p_2,...,p_6$

根据高中概率知识，假设硬币正面的概率 $p_H=0.5$，H代表正面，T代表反面，我们知道投掷一次得到正面的概率为0.5，那么投掷两次为正面的概率为0.25，如果用$p_H$表示的话，$p(HH|\theta)=p_H \times p_H$

把这个函数画出来，就是下面这个样子的。是一个顶点在原点的抛物线。



如果我们投掷三次，得到两正一反，那么此时的似然函数为：

$$p(HHT|\theta)=p_H\times p_H\times(1-p_H)$$

此时把这个函数画出来，就是这样的。是一个两端在0和1的三次函数。



不难吧，在投掷硬币的试验中，似然函数就是一系列多项式组成的。概率描述了已知参数时的随机变量的输出结果；似然则用来描述已知随机变量输出结果时，未知参数的可能取值，而在上图中，$p_H=0.66$ 时似然函数取最大值，就代表说从现有的数据中看，$p_H=0.66$ 是看起来最“似然”的～

### 极大似然估计：

有了似然函数，我们就可以用似然函数来估计模型的参数了。根据直觉，看起来最似然的地方应该就是参数最合理的估计，转化为数学语言是我们要估计的模型的参数就是使得似然函数取极大值的参数。

假设投掷硬币100次，正面53次，则似然函数可以表达为：

$$p(D|\theta)=p_H^{53}\times(1-p_H)^{49}$$

我们知道，一阶导数为0时函数取最大值，则

$$\frac{dp(D|\theta)}{dp_H}=53p_H^{52}\times(1-p_H)^47-47p_H^{53}\times(1-p_H)^46=(\frac{53}{p_H}-\frac{47}{1-p_H})(p_H^{53}(1-p_H)^{47})=0$$

因此，$\frac{53}{p_H}-\frac{47}{1-p_H}=0\rightarrow p_H=0.53$

这样我们就根据实验结果，投掷硬币100次，正面53次，估计出了模型的参数 $p_H=0.53$。

把极大似然估计推广到更一般的情况：价值投掷硬币 $n$ 次，正面 $n_1$。

第一步写出似然函数：

$$p(D|\theta)=p_H^{n_1}\times (1-p_H)^{n-n_1}$$

第二步，取对数log，因为对数函数单调递增的性质，似然函数的极大值与取对数后的似然函数的极大值是一样的，但是对于求解更方便：

$$\text{log}(p(D|\theta))=n_1\text{log}(p_H)+(n-n_1)\text{log}(1-p_H)$$

第三步，求导数，并令导数为0：

$$\frac{d\text{log}(p(D|\theta))}{dp_H}=\frac{n_1}{p_H}-\frac{n-n_1}{1-p_H}=0$$

所以 

$$\frac{n_1}{p_H}=\frac{n-n_1}{1-p_H}\rightarrow p_H=\frac{n_1}{n}$$

根据最大似然估计，硬币正面的概率，就是正面出现的概率除以总的投掷次数，换句话说，就是正面出现的频率～这就是频率学派眼中估计概率的方式。

可是这样估计，如果我只投掷了1次，正面出现1次，按照最大似然估计我硬币正面的概率就是1，这是明显有问题的。

这时贝叶斯学派站出来了，说我们觉得这样估计参数有问题，还应该考虑一些在做实验之前的经验，例如这枚硬币如果没被动过手脚，那么其正面的概率应该在0.5附近。

贝叶斯学派认为应该把这些做实验之前的经验（先验概率）加入到参数估计的过程中，于是提出了最大后验估计～
