### tf.truncated_normal_initializer

#### 1. What is different between tf.truncated_normal and tf.random_normal?

`tf.random_normal_initializer(shape, mean=0.0, stddev=1.0, dtype=tf.float32, seed=None, name=None)` outputs random values from a normal distribution.

`tf.truncated_normal_initializer(shape, mean=0.0, stddev=1.0, dtype=tf.float32, seed=None, name=None)` outputs random values from a truncated normal distribution.



The point for using truncated(截断) normal is to overcome saturation of tome functions like sigmoid (where if the value is too big/small, the neuron stops learning).
